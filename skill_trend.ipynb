```python
import pandas as pd
import numpy as np
import ast
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('data_jobs.csv')
```

```python
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
```

```python
df.info()
```

```python
df_da_sl = df[(df['job_title']== 'Data Analyst') & (df['job_country'] =='Sri Lanka')].copy()
```

```python
df_da_sl
```

```python
df_da_sl['job_post_month']=df_da_sl['job_posted_date'].dt.month
```

```python
df_da_sl_exp=df_da_sl.explode('job_skills')
df_da_sl_exp
```

```python
df_da_sl_pivot=df_da_sl_exp.pivot_table(index='job_post_month',columns='job_skills',aggfunc ='size', fill_value=0)
```

```python
df_da_sl_pivot.loc['total']=df_da_sl_pivot.sum()
df_da_sl_pivot
df_da_sl_pivot=df_da_sl_pivot[df_da_sl_pivot.loc['total'].sort_values(ascending=False).index]
```

```python
df_da_sl_pivot=df_da_sl_pivot.drop('total')
```

```python
da_total=df_da_sl.groupby('job_post_month').size()
```

```python
df_da_sl_perc=df_da_sl_pivot.div(da_total/100, axis=0)
df_da_sl_perc
```

```python
df_plot=df_da_sl_perc.iloc[:,:5]
sns.lineplot(data=df_plot, dashes=False)
plt.show()
plt.savefig('skill trend.png')
```
