```python
import pandas as pd
import numpy as np
import ast
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('data_jobs.csv')
```

```python
df.head()
```

```python
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
```

```python
df.info()
```

```python
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
```

```python
df['job_skills']
```

```python
df_sl = df[df['job_country'] =='Sri Lanka']
```

```python
df_sl.head()
```

```python
df_skills = df_sl.explode('job_skills')
df_skills
```

```python
df_skills_count = df_skills.groupby(['job_skills', 'job_title_short']).size()

df_skills_count=df_skills_count.reset_index(name='skill_count')
df_skills_count.sort_values(by = 'skill_count', ascending =False , inplace = True)
df_skills_count
```

```python
type(df_skills_count)
```

```python
print(df_skills_count.columns)
```

```python
job_titles= df_skills_count['job_title_short'].unique().tolist()
job_titles= sorted(job_titles[:3])
job_titles
```

```python
f, ax = plt.subplots(len(job_titles), 1)
for i, job_title in enumerate(job_titles):
    df_plot = df_skills_count[df_skills_count['job_title_short'] == job_title].head(5)
    df_plot.plot(kind='barh', x='job_skills', y='skill_count', ax=ax[i], title=job_title)
    ax[i].invert_yaxis()
    ax[i].set_ylabel('')
    

f.suptitle('Visualize top skills',fontsize=10)
f.tight_layout(h_pad=0.5)
plt.show()
```

```python
df_job_title_count=df_sl['job_title_short'].value_counts().reset_index(name='job_counts')
```

```python
df_skills_percentage =pd.merge(df_skills_count, df_job_title_count, how='left', on='job_title_short')
```

```python
df_skills_percentage['skill_percent']=df_skills_percentage['skill_count']/df_skills_percentage['job_counts']*100
df_skills_percentage
```

```python
f, ax = plt.subplots(len(job_titles), 1)
for i, job_title in enumerate(job_titles):
    df_plot = df_skills_percentage[df_skills_percentage['job_title_short'] == job_title].head(5)
    sns.barplot(data=df_plot, x='skill_percent', y='job_skills', ax=ax[i], hue='skill_count',palette='dark:b_r')
    ax[i].set_ylabel('')
    ax[i].set_xlabel('')
    ax[i].set_xlim(0,80)
    ax[i].set_title(job_title)
    ax[i].get_legend().remove()

f.suptitle('Skill Percentage top 3 job posting',fontsize=10)
f.tight_layout(h_pad=0.5)
plt.show()
```
