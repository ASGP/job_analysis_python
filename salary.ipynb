```python
import pandas as pd
import numpy as np
import ast
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('data_jobs.csv')
```

```python
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
```

```python
df_sl = df[(df['job_country']=='Sri Lanka')].dropna(subset=['salary_year_avg'])
df_sl
```

```python
job_titles= df_sl['job_title_short'].value_counts().index[:6].tolist()
job_titles
```

```python
df_sl_jobs=df_sl[df_sl['job_title_short'].isin(job_titles)]
```

```python
sns.boxplot(data=df_sl_jobs, x='salary_year_avg',y='job_title_short')
sns.set_theme(style='ticks')
```
